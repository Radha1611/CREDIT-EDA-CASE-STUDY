### Imputing the missing values

There are some columns where the value is mentioned as 'XNA' which means 'Not Available'. So we have to find the number of rows and columns and implement suitable techniques on them to fill those missing values or to delete them.

```python
# For Gender column

LA1['CODE_GENDER'].value_counts()
```

Since, Female is having the majority and only 4 rows are having NA values, we can update those columns with Gender 'F' as there will be no impact on the dataset.

```python
# replacing XNA with 'F'

LA1 = LA1.copy()
LA1.CODE_GENDER = LA1.CODE_GENDER.replace(to_replace = 'XNA', value = 'F')
LA1['CODE_GENDER'].value_counts()
```

```python
# Organization type column

LA1['ORGANIZATION_TYPE'].describe()
```

for column 'ORGANIZATION_TYPE', we have total count of 307511 rows of which 55374 rows are having 'XNA' values. Which means 18% of the column is having this values. Hence if we drop the rows of total 55374, will not have any major impact on our dataset.

```python
# dropping the null values rows present in the Organization_type column

LA1.drop( LA1[ LA1['ORGANIZATION_TYPE'] == 'XNA' ].index , inplace=True)
```

'AMT_ANNUITY', 'AMT_GOODS_PRICE', 'AMT_REQ_CREDIT_BUREAU_HOUR' columns is having very few null values rows. Imputing the missing values with median

```python
# caluclating the median value

AMT_ANNUITY_median = LA1['AMT_ANNUITY'].median()
AMT_ANNUITY_median
```

```python
# Filling missing values with median

LA1['AMT_ANNUITY'].fillna(AMT_ANNUITY_median, inplace = True)
```

```python
# Filling missing values with median

LA1['AMT_GOODS_PRICE'].fillna(AMT_GOODS_PRICE_median,inplace=True)
```

## Univariate analysis for categories

Now, doing Categorical Univariate Analysis in logarithmic scale for target=0(client with no payment difficulties)

```python
# Count plotting in logarithmic scale

def uniplot(LA1,col,title,hue =None):
    
    sns.set_style('white')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 20
    plt.rcParams['axes.titlepad'] = 10
    
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(LA1[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = LA1, x= col, order=LA1[col].value_counts().index,hue = hue, palette='BuGn') 
        
    plt.show()
```
